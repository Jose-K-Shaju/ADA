```python
#1
import timeit
import random
import matplotlib.pyplot as plt
def linear_search(l,ln,key):
    start=timeit.default_timer()
    for i in range(0,ln):
        if l[i]==key:
            end=timeit.default_timer()
            return end-start
    end=timeit.default_timer()
    return end-start
def main():
    l=[]
    time=[]
    n=[1000,2500,5000,7500,10000]
    for k in n:
        for i in range(0,k):
            l.append(random.randrange(0,100000))
        key=random.randrange(0,100000)
        t=linear_search(l,k,key)
        time.append(t)
    plt.plot(n,time,'-r',linewidth=1)
    plt.title('linear search')
    plt.xlabel('n')
    plt.ylabel('time')
    plt.show()
main()
```

```python
#2
import timeit
import random
import matplotlib.pyplot as plt
def bin_search(l,ln,key):
    start=timeit.default_timer()
    lb=0
    ub=ln-1
    while lb<ub:
        mid=(lb+ub)//2
        if l[mid]==key:
            end=timeit.default_timer()
            return end-start
        elif key>l[mid]:
            lb=mid+1
        elif key<l[mid]:
            ub=mid-1
    end=timeit.default_timer()
    return end-start
def main():
    l=[]
    time=[]
    n=[100,250,500,750,1000]
    for k in n:
        for i in range(0,k):
            l.append(random.randrange(0,100000))
        l=sorted(l)
        key=random.randrange(0,100000)
        t=bin_search(l,k,key)
        time.append(t)
    plt.plot(n,time,'-r',linewidth=1)
    plt.title("Binary search")
    plt.xlabel('n')
    plt.ylabel('time')
    plt.show()
main()
```

```python
#3
def towerofhanoi(n,from_rod,to_rod,aux_rod):
    if n==1:
        print(f"Move disk 1 from rod {from_rod} to rod {to_rod}.")
        return
    towerofhanoi(n-1,from_rod,aux_rod,to_rod)
    print(f"Move disk {n} from rod {from_rod} to rod {to_rod}.")
    towerofhanoi(n-1,aux_rod,to_rod,from_rod)
def main():
    n=int(input("Enter the number of disks: "))
    towerofhanoi(n,'A','C','B')
main()
```

```python
#4
import random
import timeit
import matplotlib.pyplot as plt
def selection_sort(l,ln):
    start=timeit.default_timer()
    for i in range(ln-1):
        min = i
        for j in range(i+1, ln):
            if l[j] < l[min]:
                min= j
        l[i],l[min] = l[min],l[i]
    end=timeit.default_timer()
    return end-start
def main():
    n_values=[0,250,500,750,1000]
    t=[]
    l=[]
    for i in n_values:
        l=[random.randint(0,10000) for j in range(i)]
        time=selection_sort(l,i)
        t.append(time)
    plt.plot(n_values,t)
    plt.title("Selection sort")
    plt.xlabel("n_values")
    plt.ylabel("Time taken")
    plt.show()
main() 
```

```python
#6
import timeit
import random
import matplotlib.pyplot as plt
def partition(array, low, high):
    pivot = array[high]
    i = low - 1
    for j in range(low, high):
        if array[j] <= pivot:
            i = i + 1
            array[i],array[j]=array[j], array[i]
    (array[i + 1], array[high]) = (array[high], array[i + 1])
    return i + 1
def quick_sort(array,low,high):
    if low<high:
        pi=partition(array, low, high)
        quick_sort(array,low,pi-1)
        quick_sort(array,pi+1,high)
def main():
    n_values=[0,250,500,750,1000]
    t=[]
    l=[]
    for i in n_values:
        l=[random.randint(0,10000) for j in range(i)]
        start=timeit.default_timer()
        quick_sort(l,0,i-1) 
        end=timeit.default_timer()
        ti=end-start
        t.append(ti)
    plt.plot(n_values,t)
    plt.title("Quick sort")
    plt.xlabel("n_values")
    plt.ylabel("Time taken")
    plt.show()
main() 
```

```python
# 7
def binomialCoeff(n, k):
    if k>n:
        return 0
    if k==0 or k==n:
        return 1
    return binomialCoeff(n-1, k-1) + binomialCoeff(n-1, k)
def binomialCoef(n, k):
    C = [[0 for x in range(k+1)] for x in range(n+1)]
    for i in range(n+1):
        for j in range(min(i, k)+1):
            if j == 0 or j == i:
                C[i][j] = 1
            else:
                C[i][j] = C[i-1][j-1] + C[i-1][j]
    return C[n][k] 
def main():
    n =int(input("Enter the value of n:"))
    k =int(input("Enter the value of k:"))
    print("Brute force: ")
    print (f"Value of C({n},{k}) is {binomialCoeff(n, k)}")
    print("Dynamic programming: ")
    print(f"Value of C[{str(n)}][{str(k)}] is {str(binomialCoef(n, k))}")    
main()
```

```python
#9
import timeit
def brute_force(coefficients,x):
    degree=len(coefficients)-1
    result=0
    for i in range(degree+1):
        sum=coefficients[i]
        for j in range(degree-i):
            sum=sum*x
        result+=sum
    return result
def horner(coefficients,x):
    result=0
    for coefficient in coefficients:
        result=result*x+coefficient
    return result
def main():
    degree=int(input("Enter the degree of polynomial: "))
    coefficients=[]
    print("Enter the coefficients: ")
    for i in range(0,degree):
        n=int(input())
        print(n)
        coefficients.append(n)
    x=int(input("Enter the value of x: "))
    start1=timeit.default_timer()
    print("Value of polunomial in brute force method: ",brute_force(coefficients,x))
    t1=timeit.default_timer()-start1
    start2=timeit.default_timer()
    print("Value of polynomial in Horners method: ",horner(coefficients,x))
    t2=timeit.default_timer()-start2
    print(f"Time taken with brute force method: {t1}.")
    print(f"Time taken with horner algorithm: {t2}.")
if __name__=="__main__":
    main()
```

```python
#10 Write a Program to solve the string-matching problem using Boyer-Moore
#approach.
def shift_table(pattern):
    shifttable = {}
    m = len(pattern)
    for i in range(m - 1):
        shifttable[pattern[i]] = m - 1 - i
    return shifttable
def boyer_moore(text, pattern):
    n = len(text)
    m = len(pattern)
    if m > n:
        return []
    occurrences = []
    shifttable = shift_table(pattern)
    i = 0
    while i <= n - m:
        j = m - 1
        while j >= 0 and text[i + j] == pattern[j]:
            j -= 1
        if j == -1:
            occurrences.append(i)
            i += m
        else:
            char = text[i + j]
            if char in shifttable:
                shift = shifttable[char]
            else:
                shift = m
            i += shift
    return occurrences

def main():
    text = input("Enter the text: ")
    pattern =input("Enter the pattern: ")
    result = boyer_moore(text, pattern)
    print(f"Occurrences found at positions: {result}")
if __name__=="__main__":
    main()
```

```python
#11.
def prefix_table(pattern):
    m = len(pattern)
    prefixtable = [0] * m
    j = 0
    for i in range(1, m):
        if pattern[i] == pattern[j]:
            j += 1
            prefixtable[i] = j
        else:
            if j != 0:
                j = prefixtable[j - 1]
                i -= 1
            else:
                prefixtable[i] = 0
    return prefixtable
def kmp_search(text, pattern):
    n = len(text)
    m = len(pattern)
    prefixtable = prefix_table(pattern)
    i = j = 0
    occurrences = []
    while i < n:
        if pattern[j] == text[i]:
            i += 1
            j += 1
            if j == m:
                occurrences.append(i - j)
                j = prefixtable[j - 1]
        else:
            if j != 0:
                j = prefixtable[j - 1]
            else:
                i += 1
    return occurrences
def main():
    txt="ABABDABACDABABCABABCABAB"
    pat="ABABCABAB"
    result=kmp_search(txt,pat)
    print(f"Pattern found at position {result}")
if __name__=="__main__":
    main()
```

```python
#14a 
from collections import defaultdict

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = defaultdict(list)
    def add_edge(self, u, v):
        self.graph[u].append(v)
    def topological_sort_util(self, v, visited, stack):
        visited[v] = True
        for neighbor in self.graph[v]:
            if not visited[neighbor]:
                self.topological_sort_util(neighbor, visited, stack)
        stack.append(v)
    def topological_sort(self):
        visited = [False] * self.V
        stack = []

        for v in range(self.V):
            if not visited[v]:
                self.topological_sort_util(v, visited, stack)
        while stack:
            print(stack.pop(), end=" ")
def main():
    g = Graph(6)
    g.add_edge(5, 2)
    g.add_edge(5, 0)
    g.add_edge(4, 0)
    g.add_edge(4, 1)
    g.add_edge(2, 3)
    g.add_edge(3, 1)
    print("Topological Ordering:")
    g.topological_sort()
if __name__=="__main__":
    main()
```

```python
#14b
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0] * vertices for _ in range(vertices)]
    def add_edge(self, u, v):
        self.graph[u][v] = 1
    def transitive_closure(self):
        closure = [[0] * self.V for _ in range(self.V)]
        for i in range(self.V):
            for j in range(self.V):
                closure[i][j] = self.graph[i][j]
        for k in range(self.V):
            for i in range(self.V):
                for j in range(self.V):
                    closure[i][j] = closure[i][j] or (closure[i][k] and closure[k][j])
        return closure
    def print_closure(self, closure):
        print("Transitive Closure:")
        for i in range(self.V):
            for j in range(self.V):
                print(closure[i][j], end=" ")
            print()
def main():
    g = Graph(4)
    g.add_edge(0, 1)
    g.add_edge(0, 2)
    g.add_edge(1, 2)
    g.add_edge(2, 0)
    g.add_edge(2, 3)
    g.add_edge(3, 3)
    closure = g.transitive_closure()
    g.print_closure(closure)
if __name__=="__main__":
    main()
```
